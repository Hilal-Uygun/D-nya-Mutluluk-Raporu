```python
bins = [2.800, 3.836, 4.820, 5.803, 6.786, 7.800]
labels = ["Very Unhappy", "Unhappy", "Neutral", "Happy", "Very Happy"]
```

```python
df["Score_Category"] = pd.cut(df["Score"], bins = bins, labels = labels, include_lowest=True)
```

```python
df.head(30)
```

```python
df.tail(20)
```

```python
sns.scatterplot(data=df, x="GDP per capita", y="Perceptions of corruption", hue="Score_Category")
```

```python
sns.scatterplot(data=df, x="Social support", y="Healthy life expectancy", hue="Score_Category")
```
